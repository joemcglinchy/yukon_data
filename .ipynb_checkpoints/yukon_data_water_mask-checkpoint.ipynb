```python
from gbdxtools import CatalogImage, Interface
import os, sys
import numpy as np
import geopandas as gpd
from matplotlib import pyplot as plt
from scipy.misc import bytescale
import shapely
from shapely.wkt import loads
import pandas as pd

%matplotlib inline
```

```python
image_ids = ['17669959_3',      # RE 8-29-2014, alt.id 2014-08-29T223031_RE1 (SW corner)
             '17559220_1',      # RE 8-23-2014, alt.id 2014-08-23T222157_RE5 (SW corner)
             '17511071_3',      # RE 8-14-2014, alt.id 2014-08-14T223214_RE5 (SW corner)
             '10504100112AB900' # GE 08-11-2014 (NE corner) 
            ] 
```

```python
pts_shp = 'sample_pts/sample_pts.shp'
pts = gpd.read_file(pts_shp)

fig, ax = plt.subplots(1, figsize=(4.5,10))
pts.plot(ax=ax,  color=['Red', 'Green', 'Blue'], alpha=1, legend=True)
# ax.legend(loc='middle')
```

```python
# points 1,2 of 0,1,2 should intersect the GeoEeye scene
ge_scene_ps = CatalogImage(image_ids[-1], acomp=True, pansharpen=True)
ge_scene_ms = CatalogImage(image_ids[-1], acomp=True)
```

```python
bbox = pts['geometry'][2].buffer(0.0054).bounds
ge_ps_aoi = ge_scene_ps.aoi(bbox=bbox)
ge_ms_aoi = ge_scene_ms.aoi(bbox=bbox)
```

```python
ge_ps_aoi.shape, ge_ms_aoi.shape
```

```python
# %matplotlib qt
fig, ax = plt.subplots(1,2, figsize = (20,10))
titles = ('Pansharpened', 'Multispectral')
for i, data in enumerate([ge_ps_aoi, ge_ms_aoi]):
    ax[i].imshow(bytescale(np.rollaxis(data[[2,1,0], :, :], 0, 3)))
    ax[i].set_title(titles[i])
    #data.plot()
    
plt.show()
```

```python
## look at water index
ge_ms_aoi.shape
```

```python
# ndwi with green and nir
ndwi = (ge_ms_aoi[1,:,:] - ge_ms_aoi[3,:,:]) / (ge_ms_aoi[1,:,:] + ge_ms_aoi[3,:,:])
```

```python
fig, ax = plt.subplots(figsize = (20,10))
ax.imshow(ndwi)
ax.set_title('ndwi')
    #data.plot()
    
plt.show()
```

```python
plt.hist(ndwi.flatten(), bins=255)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (20,10))
ax.imshow(ndwi>0.2)
ax.set_title('ndwi')
    #data.plot()
    
plt.show()
```

## Check a WV2/3 image

```python
test_id = '103001007D0A7D00'
wv = CatalogImage(test_id)
wv_aoi = wv.aoi(bbox=bbox)
wv_aoi.shape
```

```python
## show the image
fig, ax = plt.subplots(figsize = (20,10))
ax.imshow(bytescale(np.rollaxis(wv_aoi[[4,2,1], :, :], 0, 3)))
ax.set_title('WV2')
plt.show()
```

```python
# ndwi with green and nir
wv_ndwi = (wv_aoi[2,:,:] - wv_aoi[6,:,:]) / (wv_aoi[2,:,:] + wv_aoi[6,:,:])   # green and nir1
wv_ndwi2 = (wv_aoi[0,:,:] - wv_aoi[7,:,:]) / (wv_aoi[0,:,:] + wv_aoi[7,:,:])  # coastal and nir2
```

```python
# %matplotlib qt
fig, ax = plt.subplots(1,2, figsize = (20,10))
titles = ('ndwi green', 'ndwi coastal')
for i, data in enumerate([wv_ndwi, wv_ndwi2]):
    ax[i].imshow(data)
    ax[i].set_title(titles[i])
    #data.plot()
    
plt.show()
```

```python
fig, ax = plt.subplots(1,2, figsize = (20,10))
titles = ('ndwi green', 'ndwi coastal')
for i, data in enumerate([wv_ndwi, wv_ndwi2]):
    ax[i].hist(data.flatten(), bins=255)
    ax[i].set_title(titles[i])
    #data.plot()
    
plt.show()
```

```python
fig, ax = plt.subplots(1,2, figsize = (20,10))
titles = ('ndwi green', 'ndwi coastal')
for i, tup in enumerate(zip([wv_ndwi, wv_ndwi2], (0.2, 0.6))):
    data, thresh = tup
    print(thresh)
    ax[i].imshow(data>thresh)
    ax[i].set_title(titles[i])
    #data.plot()
    
plt.show()
```

```python
wv_ndwi.shape
```

```python
from skimage.filters import threshold_otsu

thresh_otsu_coastalNDWI = threshold_otsu(wv_ndwi2.compute())
thresh_otsu_greenNDWI = threshold_otsu(wv_ndwi.compute())

fig, ax = plt.subplots(1,2, figsize = (20,10))
titles = ('ndwi green', 'ndwi coastal')
for i, tup in enumerate(zip([wv_ndwi, wv_ndwi2], (thresh_otsu_greenNDWI, thresh_otsu_coastalNDWI))):
    data,thresh = tup
    ax[i].hist(data.flatten(), bins=255)
    ax[i].axvline(x=thresh, color='r')
    ax[i].set_title(titles[i])
    #data.plot()
    
plt.show()
```

```python
thresh_otsu_greenNDWI
```

```python
fig, ax = plt.subplots(1,2, figsize = (20,10))
titles = ('ndwi green', 'ndwi coastal')
for i, tup in enumerate(zip([wv_ndwi, wv_ndwi2], (thresh_otsu_greenNDWI, thresh_otsu_coastalNDWI))):
    data, thresh = tup
    print(thresh)
    ax[i].imshow(data>thresh)
    ax[i].set_title(titles[i])
    #data.plot()
    
plt.show()
```

```python
## local otsu
from skimage.morphology import disk
from skimage.filters import rank

radius = 10
selem = disk(radius)

local_otsu = rank.otsu(wv_ndwi, selem)
```

```python
local_otsu.shape
```

```python
plt.imshow(wv_ndwi >= local_otsu)
plt.show()

plt.imshow(local_otsu)
```
